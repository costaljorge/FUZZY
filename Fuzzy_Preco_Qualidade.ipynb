```python
# Definir universo de discurso para qualidade e preço
qualidade_universe = np.arange(0, 11, 0.1)
preco_universe = np.arange(0, 11, 0.1)

# Definir funções de pertinência para qualidade
qualidade_ruim = fuzz.trimf(qualidade_universe, [0, 0, 5])
qualidade_boa = fuzz.trimf(qualidade_universe, [0, 5, 7])
qualidade_muito_boa = fuzz.trimf(qualidade_universe, [5, 7, 10])

# Função para atribuir preço com base na qualidade
def atribuir_preco(qualidade_value):
    if qualidade_value <= 5:
        return 10  # Preço inicial
    elif qualidade_value <= 7:
        return 7
    else:
        return 5

# Calcular preços para cada valor de qualidade
precos = [atribuir_preco(qualidade) for qualidade in qualidade_universe]

# Plotar a relação entre qualidade e preço
plt.figure(figsize=(8, 5))
plt.plot(qualidade_universe, precos, label='Preço')
plt.title('Preço em função da Qualidade')
plt.xlabel('Qualidade')
plt.ylabel('Preço')
plt.grid(True)
plt.show()
```
